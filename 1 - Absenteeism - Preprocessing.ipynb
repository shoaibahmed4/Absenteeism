```python
column_names_reoredered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reoredered]
df.head()
```

```python
df_reason_mod = df.copy()
df_reason_mod
```

```python
type(df_reason_mod['Date'])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

```python
type(df_reason_mod['Date'])
```

```python
df_reason_mod.shape[1]
```

```python
list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod
```

```python
#list_weekday = []
#for i in range(df_reason_mod.shape[0]):
#    list_weekday.append(df_reason_mod['Date'][i].weekday())
#list_weekday
# Above method does the same as below one! above one use for loop for iteration and lower one use .apply function which is iteself iterative function.

def date_to_weekday(date_value):
    return date_value.weekday()
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis = 1)
df_reason_mod.columns.values
```

```python
column_names_updated = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 'Month Value',
       'Day of the Week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_mod = df_reason_mod[column_names_updated]
df_reason_mod
```
